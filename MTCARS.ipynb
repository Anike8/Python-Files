```python
import seaborn as sns
```

```python
tips= sns.load_dataset('tips')
```

```python
tips
```

# Stripplot

```python
sns.stripplot(y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='smoker',jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

# Swarmplot

```python
sns.swarmplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(x='tip',y='day',data=tips,hue='sex',orient='h')
plt.ylabel('tip($)')
plt.show()
```

# Boxplot

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
```

```python
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')

plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips,inner=None,color='lightgrey')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```
